```python
import numpy as np 
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup
```

```python
brand = []
battery = []
back_camera = []

color = []
display_size_cm = []
display_type = []

front_camera = []
guaranty_for_phone_year = []
guaranty_for_accessories_months = []

model = []
price = []
processor = []

reviews = []
ratings=[]
ram = []
rom = []
```

```python
for i in range(1,5):
    url = f"https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.price_range.from%3D10000&p%5B%5D=facets.price_range.to%3DMax&page={i}"
    page = requests.get(url)
    page_soup = BeautifulSoup(page.text)

    for i in page_soup.find_all("div",class_="KzDlHZ"):
        brand_name = re.findall(r"^\w+",i.text)
        if brand_name:
            brand.append(brand_name[0])
        else:
            brand.append(np.nan)
        model_name = re.findall(r"^\w+\s?(\w.+)\s?\(",i.text)
        if model_name :
            model.append(model_name[0])
        else:
            model.append(np.nan)
        color_name = re.findall(r"\((\w.+),",i.text)
        if color_name:
            color.append(color_name[0])
        else:
            color.append(np.nan)
    for v in page_soup.find_all("div",class_="Nx9bqj _4b5DiR"):
        value = re.findall(r"₹(.+)",v.text)
        if value:
            price.append(value[0])
    for i in page_soup.find_all("div",class_="tUxRFH"):
        ram_name = re.findall(r"(\d+)\s?GB\s?RAM",i.text)
        if ram_name:
            ram.append(ram_name[0])
        else:
            ram.append(np.nan)
        rom_name = re.findall(r"(\d+)\s?GB\s?ROM",i.text)
        if rom_name:
            rom.append(rom_name[0])
        else:
            rom.append(np.nan)
        back_camera_name = re.findall(r"Display(\w.+)\s?\|",i.text)
        if back_camera_name:
            back_camera.append(back_camera_name[0])
        else:
            back_camera.append(np.nan)
        front_camera_name = re.findall(r"(\d+)MP\s?Front",i.text)
        if front_camera_name:
            front_camera.append(front_camera_name[0])
        else:
            front_camera.append(np.nan)
        display_cm = re.findall(r"(\d+)\s?\s?cm",i.text)
        if display_cm:
            display_size_cm.append(display_cm[0])
        else:
            display_size_cm.append(np.nan)
        display_name = re.findall(r"inch\)\s(.+)\s?Display",i.text)
        if display_name:
            display_type.append(display_name[0])
        else:
            display_type.append(np.nan)
        guaranty_phn = re.findall(r"Processor\s?(\d+)\s",i.text)
        if guaranty_phn:
            guaranty_for_phone_year.append(guaranty_phn[0])
        else:
            guaranty_for_phone_year.append(np.nan)
        guaranty_accessories_months = re.findall(r"and\s(\d+\s*\w+)", i.text)
        if guaranty_accessories_months:
            text = guaranty_accessories_months[0].lower().strip()
            number = re.findall(r"\d+", text)
            if number:
                num = int(number[0])
                # Convert years to months
                if "year" in text:
                    guaranty_for_accessories_months.append(num * 12)
                elif "month" in text:
                    guaranty_for_accessories_months.append(num)
                else:
                    guaranty_for_accessories_months.append(np.nan)
            else:
                guaranty_for_accessories_months.append(np.nan)
        else:
            guaranty_for_accessories_months.append(np.nan)
            
        reviews_text = re.findall(r"(\d+[,\d]*)\s*Reviews", i.text)
        ratings_text = re.findall(r"(\d+[,\d]*)\s*Ratings", i.text)
        if ratings_text:
            ratings.append(ratings_text[0])
        else:
            ratings.append(np.nan)
        if reviews_text:
            reviews.append(reviews_text[0])
        else:
            reviews.append(np.nan)
    
        battery_capacity = re.findall(r"(\d+)\smAh\sBattery",i.text)
        if battery_capacity:
            battery.append(battery_capacity[0])
        else:
            battery.append(np.nan)
        value = re.findall(r"Battery(.+)\sProcessor",i.text)
        if value:
            processor.append(value[0])
        else:
            value = re.findall(r"Camera(.+)\sProcessor\sProcessor",i.text)
            if value:
                processor.append(value[0])
            else:
                processor.append(np.nan)
dictionary = {"Brand":brand,"Model":model,"Color":color,"Ram":ram,"Rom":rom,"Display_size_cm":display_size_cm,"Display_Type":display_type,
              "Front Camera":front_camera,"Back Camera":back_camera,"Guaranty for phone":guaranty_for_phone_year,"Processor":processor,
              "Guaranty for Accessories months":guaranty_for_accessories_months,"Reviews":reviews,"Ratings":ratings,"Battery":battery,"Price":price}
```
